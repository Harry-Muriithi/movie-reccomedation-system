````python
# 🧱 Step 1: Write application_layer.py with poster fetching, grid display & CF predict
app_code = """
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import requests
from itertools import zip_longest
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from data_layer import load_and_prepare_data
from model_layer import load_ratings_data, CollaborativeFiltering

# --- Load & Prepare ---
df = load_and_prepare_data()       # includes df['id']
ratings_df = load_ratings_data()
cf = CollaborativeFiltering(ratings_df)

st.set_page_config(page_title="Movie Recommender", layout="wide")
st.title("🎬 Movie Recommender System")
st.success("✅ Data & CF model loaded!")

# --- Constants ---
TMDB_IMAGE_BASE  = "https://image.tmdb.org/t/p/w342"
TMDB_API_IMAGES  = "https://api.themoviedb.org/3/movie/{}/images"
HEADERS = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiJlZDgzOTQyODZhY2FmN2Y1ODg0ZGMwNGYxMzM5M2MwMyIsIm5iZiI6MTc0ODIwODkyNy4zNTM5OTk5LCJzdWIiOiI2ODMzOGQxZjExZGNlZTg3MzU4MzdiY2IiLCJzY29wZXMiOlsiYXBpX3JlYWQiXSwidmVyc2lvbiI6MX0.NaDGWTV0eWKe7mG1qJs38RxYmgaFEsuRjx4yEipcU_k "
}
PLACEHOLDER = "https://via.placeholder.com/342x513?text=No+Image"
DEFAULT_COLS = 5

@st.cache_data(show_spinner=False)
def fetch_poster(movie_id):
    try:
        resp = requests.get(TMDB_API_IMAGES.format(movie_id), headers=HEADERS, timeout=5)
        posters = resp.json().get('posters', [])
        if posters:
            return TMDB_IMAGE_BASE + posters[0]['file_path']
    except:
        pass
    return PLACEHOLDER

def show_movies_grid(rows, cols=DEFAULT_COLS):
    for grp in zip_longest(*[iter(rows)]*cols, fillvalue=None):
        cols_layout = st.columns(cols)
        for cell, col in zip(grp, cols_layout):
            if not cell: continue
            _, row = cell
            img = fetch_poster(row['id'])
            col.image(img, width=180)
            col.markdown(
                f"<p style='text-align:center;font-weight:bold;'>{row['title']}</p>",
                unsafe_allow_html=True
            )

st.markdown("---")

# 1) Popularity Chart
with st.expander("📊 Top Movies by Popularity", expanded=True):
    top = df.nlargest(10, 'popularity')[['title','popularity']]
    fig, ax = plt.subplots(figsize=(10,6))
    sns.barplot(data=top, y='title', x='popularity', palette='rocket', ax=ax)
    ax.set_title("Top 10 Movies by Popularity")
    st.pyplot(fig)

st.markdown("---")

# 2) Demographic-Based Recs
with st.expander("🎯 Demographic Recs", expanded=False):
    m = df['vote_count'].quantile(0.9)
    C = df['vote_average'].mean()
    kval = df[df['vote_count']>=m].copy()
    kval['score'] = kval.apply(
        lambda x: (x['vote_count']/(x['vote_count']+m)*x['vote_average'])
                  + (m/(m+x['vote_count'])*C), axis=1
    )
    if st.button("Show 5 Demographic Recs"):
        show_movies_grid(kval.sort_values('score', ascending=False).sample(5).iterrows())

st.markdown("---")

# 3) Genre-Based Recs
with st.expander("🔍 Genre-Based Recs", expanded=False):
    choice = st.selectbox("Choose a Movie:", df['title'], key='genre')
    if st.button("Get Genre Recs"):
        sel = df[df['title']==choice]
        genres = [g['name'] for g in ast.literal_eval(sel.iloc[0]['genres'])]
        mask = df['genres'].apply(
            lambda s: any(name in [x['name'] for x in ast.literal_eval(s)] for name in genres)
        )
        recs = df[mask & (df['title']!=choice)].head(5)
        show_movies_grid(recs.iterrows())

st.markdown("---")

# 4) Plot-Based Recs
with st.expander("🔍 Plot-Based Recs", expanded=False):
    tfidf = TfidfVectorizer(stop_words='english')
    mat   = tfidf.fit_transform(df['overview'].fillna(''))
    sim   = cosine_similarity(mat, mat)
    idx   = pd.Series(df.index, index=df['title']).drop_duplicates()
    pmovie = st.selectbox("Choose a Movie:", df['title'], key='plot')
    if st.button("Get Plot Recs"):
        i = idx[pmovie]
        vals = sorted(enumerate(sim[i]), key=lambda x: x[1], reverse=True)[1:6]
        show_movies_grid(df.iloc[[v for v,_ in vals]].iterrows())

st.markdown("---")

# 5) Metadata-Based Recs
with st.expander("🔍 Metadata-Based Recs", expanded=False):
    # parse JSON columns
    for col in ['cast','crew','keywords','genres']:
        df[col] = df[col].apply(ast.literal_eval)
    df['cast_names']    = df['cast'].apply(lambda lst: [d['name'] for d in lst][:3])
    df['keyword_names'] = df['keywords'].apply(lambda lst: [d['name'] for d in lst][:3])
    df['genre_names']   = df['genres'].apply(lambda lst: [d['name'] for d in lst][:3])
    df['director']      = df['crew'].apply(
        lambda lst: next((d['name'] for d in lst if d['job']=="Director"), "")
    )
    # build soup
    df['soup'] = df.apply(
        lambda x: " ".join(x['cast_names'] + x['keyword_names'] + [x['director']] + x['genre_names']),
        axis=1
    )
    mat2 = CountVectorizer(stop_words='english').fit_transform(df['soup'])
    sim2 = cosine_similarity(mat2, mat2)
    idx2 = pd.Series(df.index, index=df['title']).drop_duplicates()
    mchoice = st.selectbox("Choose a Movie:", df['title'], key='meta')
    if st.button("Get Meta Recs"):
        i = idx2[mchoice]
        vals = sorted(enumerate(sim2[i]), key=lambda x: x[1], reverse=True)[1:6]
        show_movies_grid(df.iloc[[v for v,_ in vals]].iterrows())

st.markdown("---")

# 6) Collaborative Filtering Prediction
with st.expander("🤝 CF Rating Prediction", expanded=False):
    uid = st.number_input("User ID", min_value=1, step=1)
    mid = st.number_input("Movie ID", min_value=1, step=1)
    if st.button("Predict Rating"):
        est = cf.predict_rating(uid, mid)
        st.write(f"**Predicted rating:** {est:.2f}")
"""
with open("application_layer.py","w") as f:
    f.write(app_code)


# 🧱 Step 2: Write data_layer.py
data_layer_code = """
import pandas as pd, kagglehub, os

def load_and_prepare_data():
    path = kagglehub.dataset_download("chaitanyasood1/tmdb-5000-movies")
    mv = pd.read_csv(os.path.join(path, "tmdb_5000_movies.csv"))
    cr = pd.read_csv(os.path.join(path, "tmdb_5000_credits.csv"))
    df = pd.merge(mv, cr, on='title').dropna().drop_duplicates()
    for c in ['genres','cast','crew','keywords']:
        df[c] = df[c].astype(str)
    df['overview']     = df['overview'].fillna('')
    df['vote_count']   = pd.to_numeric(df['vote_count'], errors='coerce')
    df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
    df['popularity']   = pd.to_numeric(df['popularity'], errors='coerce')
    return df
"""
with open("data_layer.py","w") as f:
    f.write(data_layer_code)


# 🧱 Step 3: Write model_layer.py
model_layer_code = """
import pandas as pd, kagglehub, os
from surprise import Dataset, Reader, SVD

def load_ratings_data():
    path = kagglehub.dataset_download("grouplens/movielens-latest-small")
    return pd.read_csv(os.path.join(path, "ratings.csv"))

class CollaborativeFiltering:
    def __init__(self, ratings_df):
        reader = Reader(rating_scale=(0.5,5))
        data = Dataset.load_from_df(ratings_df[['userId','movieId','rating']], reader)
        trainset = data.build_full_trainset()
        self.model = SVD()
        self.model.fit(trainset)

    def predict_rating(self, uid, mid):
        return self.model.predict(uid, mid).est
"""
with open("model_layer.py","w") as f:
    f.write(model_layer_code)


# 🧱 Step 4: Install dependencies & launch via ngrok
!pip install streamlit pyngrok kagglehub scikit-learn scikit-surprise seaborn matplotlib -q
from pyngrok import ngrok
import os

ngrok.kill()
os.system("streamlit run application_layer.py &")
url = ngrok.connect(8501).public_url
print(f"✅ App live at: {url}")
````
